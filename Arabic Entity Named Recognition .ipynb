```python
dataset='C:\\Users\\Yara\\Desktop\\AQMAR DATASET'
file_paths=[]
for file in os.listdir(dataset):
    if file.endswith(".txt"):
        file_path =dataset+'\\'+file
        file_paths.append(file_path)
```

```python
def read_text_file(files_paths,files_content):
    files_content=""
    for file_path in files_paths:
        with open(file_path,encoding="utf8") as f:
            text=f.read()
        files_content+=text
    return(files_content)
```

```python
all_contents=[]
print(len(file_paths))
all_files=read_text_file(file_paths,all_contents)
print(len(all_files))
```

```python
all_sentences = nltk.sent_tokenize(all_files)
T=[]
for i in all_sentences:
    processed_article = re.sub(r'\\s+', '. ', i)
    text = re.sub("[إأآا]", "ا", processed_article)
    text = re.sub("ى", "ي", text)
    text = re.sub("ؤ", "ء", text)
    text = re.sub("ئ", "ء", text)
    text = re.sub("ة", "ه", text)
    text = re.sub("گ", "ك", text)
    text = re.sub(r'[\u064B-\u0652]', '', text)
    text = text.replace('وو', 'و')
    text = text.replace('يي', 'ي')
    text = text.replace('اا', 'ا')
    p_tashkeel = re.compile(r'[\u0617-\u061A\u064B-\u0652]')
    text = re.sub(p_tashkeel,"", text)
    p_longation = re.compile(r'(.)\1+')
    subst = r"\1\1"
    text = re.sub(p_longation, subst, text)
    text = text.strip()
    T.append(text)

print(T)
```

```python
LABELS=['O','B-PERS','I-PERS','B-DATE','I-DATE','B-ORDINAL','I-ORDINAL','B-GPE','I-GPE','B-NORP','B-MIS1','I-PER','B-PER','B-MIS2','I-MIS2','B-MIS3',
            'I-NORP','B-WEBSITE','I-WEBSITE','B-CARDINAL','I-CARDINAL','B-OCC','I-OCC','B-EVENT','B-MIS0','I-MIS0','I-MIS1',
            'I-EVENT','B-LOC','I-LOC','B-FAC','I-FAC','B-MONEY','I-MONEY','B-TIME','I-TIME','B-LANGUAGE','I-LANGUAGE',
            'B-UNIT','B-PRODUCT','I-PRODUCT','B-CURR','I-CURR','B-PERCENT','I-PERCENT','B-LAW','I-LAW','B-QUANTITY','I-QUANTITY','B-ORG','I-ORG']
words = [nltk.word_tokenize(sent) for sent in T]
```

```python
for i in range(len(words)):
    if words[i][0] in LABELS:
        words[i].remove(words[i][0])
    else:
        continue
```

```python
vocab=set()
def get_label_and_data(words):
    sentence_label=[]
    sentance_data=[]
    for i in words:
        y=[]
        x=[]
        for j in i:
           if j in LABELS:
                y.append(j)
           else:
               if j=='.':
                   continue
               else:
                   x.append(j)
                   vocab.add(j)
        sentence_label.append(y)
        sentance_data.append(x)
    return sentance_data,sentence_label
```

```python
from tensorflow.keras.utils import to_categorical
def encode_labels(entitys_sentences,tag2id):
    encoded_labels=[]
    for i in entitys_sentences:
        encoded_tags = [tag2id[tag] for tag in i]
        new_tags = to_categorical(encoded_tags, num_classes=51)
        encoded_labels.append(np.array(new_tags))
    return encoded_labels
```

```python
def lstm_predict(sentence,lstm_model,word2vec2,tags_onehot):
    words = [nltk.word_tokenize(sentence)]
    word_vectors = [np.array(word2vec2[w]) for w in words[0] if w in word2vec2] # Get vectors for words in model
    word_vectors=np.array(word_vectors)
    max_seq_len = 40
    vector_dim = 300  # Dimensionality of Word2Vec embeddings
    padded_sequences = pad_sequences(
        [word_vectors],
        maxlen=max_seq_len,
        dtype='float32',
        padding='post',     # Pad after the sequence
        truncating='post',  # Truncate if sequence is longer than max_seq_len
        value=np.zeros(vector_dim)  # Pad with zero vectors (300-dimensional)
    )
   
    predictions = lstm_model.predict(padded_sequences)
   
    def get_key(val):
        for key, value in tag2one_hote.items():
            if np.argmax(value) == val:
                return key
        return None  # In case no match is found
    # Assuming pred contains the prediction output for each token in the sentence
    for i in range(len(words[0])):
        predicted_index = np.argmax(predictions[0][i])  # Get the predicted label index for each token
        entity_label = get_key(predicted_index)   # Get the corresponding entity label from tag2one_hote
        print( words[0][i],  entity_label)
```
